**Methodology: A Multi-Model Ensemble Approach**

Our forecast utilizes an ensemble method, combining the strengths of three distinct time series models:

*   **ARIMA (AutoRegressive Integrated Moving Average):** A statistical model capturing temporal dependencies and seasonality.
*   **Prophet:** Developed by Meta, designed for time series with strong seasonal effects and historical data gaps.
*   **XGBoost:** A machine learning gradient boosting framework, adapted here to use lagged delivery data as features.

By averaging the forecasts from these diverse models, the ensemble mean aims to provide a more robust prediction, mitigating the risks associated with relying on any single model.

**Historical Performance Check: Which Model Was Most Accurate?**

Before projecting the future, it's crucial to evaluate how well these models performed on past data. Backtesting was conducted on the last 8 quarters (2023Q1 - 2024Q4), comparing each model's predictions against actual reported deliveries.

*Insight:* **ARIMA** demonstrated the strongest historical accuracy on this dataset, with the lowest average error in predicting past quarters. While Prophet had larger errors in this specific backtest window, both ARIMA and XGBoost proved relatively closer to actual outcomes historically.

```python
# Then run the forecasting code
import pandas as pd, numpy as np, os, warnings, math
warnings.filterwarnings("ignore")
```

```python
# If running in Colab, upload the CSV in the files pane or mount Google Drive; otherwise ensure CSV path is correct.
CSV_PATH = "tesla_quarterly_2013Q1_2025Q3.csv"  # upload this file into Colab workspace
```

```python
# Load CSV
df = pd.read_csv(CSV_PATH)
def quarter_end(year, q):
    q = int(q)
    if q==1: return f"{year}-03-31"
    if q==2: return f"{year}-06-30"
    if q==3: return f"{year}-09-30"
    return f"{year}-12-31"
df['quarter_end'] = df.apply(lambda r: quarter_end(r['year'], r['quarter_num']), axis=1)
df['ds'] = pd.to_datetime(df['quarter_end'])
df = df.sort_values('ds').reset_index(drop=True)

ts = df.set_index('ds')['deliveries'].astype(float).asfreq('Q')
ts = ts.interpolate().ffill().bfill()
```

```python
# Forecast horizon (Q4 2025 + next 4 = 5 quarters total)
h = 5
last_date = ts.index[-1]
future_index = pd.date_range(start=last_date + pd.offsets.QuarterBegin(), periods=h, freq='Q')

OUT_DIR = "/content/tesla_forecasts_outputs"
os.makedirs(OUT_DIR, exist_ok=True)
PLOT_DIR = os.path.join(OUT_DIR, "plots"); os.makedirs(PLOT_DIR, exist_ok=True)
```

```python
# 1) ARIMA (fast default)
from statsmodels.tsa.arima.model import ARIMA
arima_model = ARIMA(ts, order=(1,1,1))
arima_res = arima_model.fit()
arima_fc = arima_res.get_forecast(steps=h)
arima_mean = arima_fc.predicted_mean; arima_ci = arima_fc.conf_int(alpha=0.05)
arima_mean.index = future_index; arima_ci.index = future_index
```

```python
# 2) Prophet (optional; Colab usually supports it)
try:
    from prophet import Prophet
    df_prophet = df[['ds','deliveries']].rename(columns={'deliveries':'y'})[['ds','y']]
    m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    m.fit(df_prophet)
    future = m.make_future_dataframe(periods=h, freq='Q')
    forecast = m.predict(future)
    prop_preds = forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(h).set_index('ds')
    prop_preds = prop_preds.reindex(future_index)
    prophet_ok = True
except Exception as e:
    prop_preds = None
    prophet_ok = False
```

```python
# 3) XGBoost (lag features)
try:
    import xgboost as xgb
    lags = 6
    df_l = pd.DataFrame({'y': ts.values}, index=ts.index)
    for lag in range(1, lags+1):
        df_l[f'lag_{lag}'] = df_l['y'].shift(lag)
    df_l = df_l.dropna()
    X = df_l.drop(columns='y').values; y = df_l['y'].values
    model_xgb = xgb.XGBRegressor(n_estimators=100, max_depth=3, objective='reg:squarederror', verbosity=0)
    model_xgb.fit(X, y)
    history = list(ts.values); xgb_preds = []
    for i in range(h):
        last_vals = history[-lags:]
        if len(last_vals) < lags:
            last_vals = [np.nan]*(lags - len(last_vals)) + last_vals
        Xnew = np.array(last_vals[::-1]).reshape(1, -1)
        yhat = float(model_xgb.predict(Xnew)[0])
        xgb_preds.append(yhat); history.append(yhat)
    xgb_series = pd.Series(xgb_preds, index=future_index)
    resid_std = float(np.std(y - model_xgb.predict(X)))
    xgb_ci_lower = xgb_series - 1.96*resid_std; xgb_ci_upper = xgb_series + 1.96*resid_std
    xgb_ok = True
except Exception as e:
    xgb_series = None; xgb_ci_lower = None; xgb_ci_upper = None; xgb_ok = False
```

```python
# Build forecast dataframe
fc = pd.DataFrame(index=future_index)
fc['arima_mean'] = arima_mean.values; fc['arima_lower'] = arima_ci.iloc[:,0].values; fc['arima_upper'] = arima_ci.iloc[:,1].values

if prophet_ok:
    fc['prophet_mean'] = prop_preds['yhat'].values; fc['prophet_lower'] = prop_preds['yhat_lower'].values; fc['prophet_upper'] = prop_preds['yhat_upper'].values
else:
    fc['prophet_mean'] = [np.nan]*h; fc['prophet_lower'] = [np.nan]*h; fc['prophet_upper'] = [np.nan]*h

if xgb_ok:
    fc['xgb_mean'] = xgb_series.values; fc['xgb_lower'] = xgb_ci_lower.values; fc['xgb_upper'] = xgb_ci_upper.values
else:
    fc['xgb_mean'] = [np.nan]*h; fc['xgb_lower'] = [np.nan]*h; fc['xgb_upper'] = [np.nan]*h

fc['ensemble_mean'] = fc[['arima_mean','prophet_mean','xgb_mean']].mean(axis=1, skipna=True)
fc['ensemble_lower'] = fc[['arima_lower','prophet_lower','xgb_lower']].min(axis=1, skipna=True)
fc['ensemble_upper'] = fc[['arima_upper','prophet_upper','xgb_upper']].max(axis=1, skipna=True)
```

```python
# Save CSV + plots
fc.to_csv(os.path.join(OUT_DIR, "tesla_forecasts_5q.csv"), index_label='quarter_end')
df[['ds','deliveries']].set_index('ds').to_csv(os.path.join(OUT_DIR, "tesla_historical.csv"))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,4))
plt.plot(ts.index, ts.values, marker='o', label='Historical'); plt.title('Historical deliveries'); plt.grid(True); plt.legend(); plt.tight_layout(); plt.savefig(os.path.join(PLOT_DIR,'historical.png')); plt.close()

plt.figure(figsize=(9,4))
plt.plot(ts.index, ts.values, marker='o', label='Historical'); plt.plot(fc.index, fc['arima_mean'], marker='o', linestyle='--', label='ARIMA'); plt.fill_between(fc.index, fc['arima_lower'], fc['arima_upper'], alpha=0.2); plt.title('ARIMA forecast'); plt.grid(True); plt.legend(); plt.tight_layout(); plt.savefig(os.path.join(PLOT_DIR,'arima.png')); plt.close()

if prophet_ok:
    plt.figure(figsize=(9,4))
    plt.plot(ts.index, ts.values, marker='o', label='Historical'); plt.plot(fc.index, fc['prophet_mean'], marker='o', linestyle='--', label='Prophet'); plt.fill_between(fc.index, fc['prophet_lower'], fc['prophet_upper'], alpha=0.2); plt.title('Prophet forecast'); plt.grid(True); plt.legend(); plt.tight_layout(); plt.savefig(os.path.join(PLOT_DIR,'prophet.png')); plt.close()

if xgb_ok:
    plt.figure(figsize=(9,4))
    plt.plot(ts.index, ts.values, marker='o', label='Historical'); plt.plot(fc.index, fc['xgb_mean'], marker='o', linestyle='--', label='XGBoost'); plt.fill_between(fc.index, fc['xgb_lower'], fc['xgb_upper'], alpha=0.2); plt.title('XGBoost forecast'); plt.grid(True); plt.legend(); plt.tight_layout(); plt.savefig(os.path.join(PLOT_DIR,'xgb.png')); plt.close()

plt.figure(figsize=(9,4))
plt.plot(ts.index, ts.values, marker='o', label='Historical'); plt.plot(fc.index, fc['ensemble_mean'], marker='o', linestyle='--', label='Ensemble');
if not fc['ensemble_lower'].isna().all() and not fc['ensemble_upper'].isna().all():
    plt.fill_between(fc.index, fc['ensemble_lower'], fc['ensemble_upper'], alpha=0.2)
plt.title('Ensemble forecast'); plt.grid(True); plt.legend(); plt.tight_layout(); plt.savefig(os.path.join(PLOT_DIR,'ensemble.png')); plt.close()

print("Outputs saved to:", OUT_DIR)
print("\nEnsemble forecasts (next 5 quarters):")
print(fc[['ensemble_mean','ensemble_lower','ensemble_upper']])
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import xgboost as xgb
import numpy as np

# Load your Tesla deliveries dataset
df = pd.read_csv("tesla_quarterly_2013Q1_2025Q3.csv")

# Function to determine quarter end date
def quarter_end(year, q):
    q = int(q)
    if q==1: return f"{year}-03-31"
    if q==2: return f"{year}-06-30"
    if q==3: return f"{year}-09-30"
    return f"{year}-12-31"

# Ensure proper datetime for Prophet/ARIMA
df['ds'] = df.apply(lambda r: quarter_end(r['year'], r['quarter_num']), axis=1)
df['ds'] = pd.to_datetime(df['ds'])
df = df.rename(columns={'deliveries':'y'})[['ds','y']]


# --------------------
# Prophet Forecast
# --------------------
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(periods=4, freq='Q')  # Forecast 1 year ahead
prophet_forecast = m.predict(future)[['ds','yhat','yhat_lower','yhat_upper']].tail(4)

# --------------------
# ARIMA Forecast
# --------------------
arima = ARIMA(df['y'], order=(2,1,2))
arima_fit = arima.fit()
arima_forecast = arima_fit.get_forecast(steps=4)
arima_ci = arima_forecast.conf_int()
arima_df = pd.DataFrame({
    'ds': pd.date_range(df['ds'].iloc[-1] + pd.offsets.QuarterEnd(), periods=4, freq='Q'),
    'yhat': arima_forecast.predicted_mean,
    'yhat_lower': arima_ci.iloc[:,0],
    'yhat_upper': arima_ci.iloc[:,1]
})

# --------------------
# XGBoost Forecast
# --------------------
lags = 8
X, y = [], []
for i in range(lags, len(df)):
    X.append(df['y'].iloc[i-lags:i].values)
    y.append(df['y'].iloc[i])
X, y = np.array(X), np.array(y)

model = xgb.XGBRegressor(n_estimators=200, max_depth=3, learning_rate=0.05)
model.fit(X,y)

last_window = df['y'].iloc[-lags:].values
xgb_preds = []
for _ in range(4):
    pred = model.predict(last_window.reshape(1,-1))[0]
    xgb_preds.append(pred)
    last_window = np.roll(last_window,-1)
    last_window[-1] = pred

xgb_df = pd.DataFrame({
    'ds': pd.date_range(df['ds'].iloc[-1] + pd.offsets.QuarterEnd(), periods=4, freq='Q'),
    'yhat': xgb_preds
})

# --------------------
# Combine Forecasts into forecast_table
# --------------------
forecast_table = pd.DataFrame({
    'Quarter': prophet_forecast['ds'].dt.to_period("Q").astype(str),
    'Prophet': prophet_forecast['yhat'].values,
    'ARIMA': arima_df['yhat'].values,
    'XGBoost': xgb_df['yhat'].values
})
forecast_table['Optimistic'] = forecast_table[['Prophet','ARIMA','XGBoost']].max(axis=1)
forecast_table['Conservative'] = forecast_table[['Prophet','ARIMA','XGBoost']].min(axis=1)
forecast_table['Ensemble Mean'] = forecast_table[['Prophet','ARIMA','XGBoost']].mean(axis=1)

forecast_table
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from statsmodels.tsa.arima.model import ARIMA
from xgboost import XGBRegressor
import numpy as np
```

```python
# Load historical data
# ======================
df = pd.read_csv("tesla_quarterly_2013Q1_2025Q3.csv")
df["Date"] = pd.to_datetime(df["year"].astype(str) + "-" + df["quarter"].str[-2:] + "-01", format='%Y-Q%m-01')
df = df.sort_values("Date")
ts = df[["Date", "deliveries"]].rename(columns={"Date": "ds", "deliveries": "y"})
```

```python
# Prophet Forecast
# ======================
prophet = Prophet(yearly_seasonality=False, weekly_seasonality=False, daily_seasonality=False)
prophet.fit(ts)
future = prophet.make_future_dataframe(periods=5, freq="Q")
forecast_prophet = prophet.predict(future).tail(5)[["ds", "yhat"]]
```

```python
# ARIMA Forecast
# ======================
arima = ARIMA(ts["y"], order=(2,1,2))
arima_fit = arima.fit()
forecast_arima = arima_fit.forecast(steps=5)
forecast_arima = pd.DataFrame({"ds": forecast_prophet["ds"], "yhat": forecast_arima.values})
```

```python
# XGBoost Forecast
# ======================
# Prepare features (lags)
def make_lags(series, lags=4):
    df_lags = pd.DataFrame({"y": series})
    for lag in range(1, lags+1):
        df_lags[f"lag{lag}"] = df_lags["y"].shift(lag)
    return df_lags.dropna()

lags = make_lags(ts["y"], lags=4)
X, y = lags.drop("y", axis=1), lags["y"]
xgb = XGBRegressor(n_estimators=200, max_depth=3, learning_rate=0.1)
xgb.fit(X, y)
```

```python
# Forecast recursively
last_vals = list(ts["y"].values[-4:])
xgb_forecasts = []
for _ in range(5):
    X_pred = np.array(last_vals[-4:]).reshape(1, -1)
    y_pred = xgb.predict(X_pred)[0]
    xgb_forecasts.append(y_pred)
    last_vals.append(y_pred)

forecast_xgb = pd.DataFrame({"ds": forecast_prophet["ds"], "yhat": xgb_forecasts})
```

```python
# Combine Forecasts
# ======================
combined = forecast_prophet.rename(columns={"yhat":"Prophet"})
combined["ARIMA"] = forecast_arima["yhat"].values
combined["XGBoost"] = forecast_xgb["yhat"].values
combined["Ensemble"] = combined[["Prophet", "ARIMA", "XGBoost"]].mean(axis=1)

print("\nForecast Comparison (Q4 2025  Q4 2026):")
print(combined)
```

```python
# Backtesting for model accuracy
# ======================

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Choose backtesting window (e.g., last 8 quarters)
backtest_periods = 8
train_data = df.iloc[:-backtest_periods]
test_data = df.iloc[-backtest_periods:]
```

```python
# --- Prophet backtest ---
prophet_bt = Prophet()
prophet_bt.fit(train_data.rename(columns={"quarter": "ds", "deliveries": "y"}))
future_bt = prophet_bt.make_future_dataframe(periods=backtest_periods, freq="Q")
forecast_prophet_bt = prophet_bt.predict(future_bt)
prophet_pred = forecast_prophet_bt.tail(backtest_periods)["yhat"].values
```

```python
# --- ARIMA backtest ---
arima_bt = ARIMA(train_data["deliveries"], order=(2,1,2))
arima_fit_bt = arima_bt.fit()
forecast_arima_bt = arima_fit_bt.forecast(steps=backtest_periods)
arima_pred = forecast_arima_bt.values
```

```python
# --- XGBoost backtest ---
xgb_bt = XGBRegressor(objective="reg:squarederror", n_estimators=200)
x = np.arange(len(train_data)).reshape(-1, 1)
y = train_data["deliveries"].values
xgb_bt.fit(x, y)
future_x = np.arange(len(train_data), len(train_data)+backtest_periods).reshape(-1, 1)
xgb_pred = xgb_bt.predict(future_x)
```

```python
# --- Ground truth ---
y_true = test_data["deliveries"].values

# --- Error metrics ---
def metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse) # Calculate RMSE by taking the square root of MSE
    return {
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": rmse
    }

results = {
    "Prophet": metrics(y_true, prophet_pred),
    "ARIMA": metrics(y_true, arima_pred),
    "XGBoost": metrics(y_true, xgb_pred)
}

# Display results
import pandas as pd
results_df = pd.DataFrame(results).T
print("\n=== Backtest Error Metrics (last 8 quarters) ===")
print(results_df.round(2))

best_model = results_df["RMSE"].idxmin()
print(f"\n Best performing model historically: {best_model}")
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Base parameters
years = np.arange(2024, 2031)
base_deliveries = np.array([1800000, 1950000, 2100000, 2250000, 2400000, 2550000, 2700000])
asp_mean, asp_std = 47000, 3000  # average selling price
margin_mean, margin_std = 0.22, 0.03  # gross margin

# Monte Carlo simulation
n_sim = 5000
results = []

for _ in range(n_sim):
    asp = np.random.normal(asp_mean, asp_std, len(years))
    margin = np.random.normal(margin_mean, margin_std, len(years))
    deliveries = base_deliveries * np.random.normal(1, 0.05, len(years))

    revenue = deliveries * asp
    gross_profit = revenue * margin
    results.append(gross_profit)

# Convert to DataFrame
results_df = pd.DataFrame(results, columns=years)
summary = results_df.describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.95]).T
summary["5%"] = summary["5%"]
summary["Median"] = summary["50%"]
summary["95%"] = summary["95%"]
summary = summary[["5%", "Median", "95%"]]
summary
```

```python
import plotly.express as px

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=summary.index, y=summary["Median"], mode='lines+markers',
    name='Median Gross Profit', line=dict(color='blue', width=3)
))
fig.add_trace(go.Scatter(
    x=summary.index, y=summary["95%"], fill=None, mode='lines',
    line=dict(color='lightblue'), name='95th Percentile'
))
fig.add_trace(go.Scatter(
    x=summary.index, y=summary["5%"], fill='tonexty', mode='lines',
    line=dict(color='lightblue'), name='5th Percentile'
))

fig.update_layout(
    title="Monte Carlo Simulation of Tesla Gross Profit (20242030)",
    xaxis_title="Year",
    yaxis_title="Gross Profit ($ billions)",
    template="plotly_dark",
    hovermode="x unified"
)

fig.show()
```
